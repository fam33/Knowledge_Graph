## Input and Preprocessing

```python
# ===============================
# STEP 1: Input & Preprocessing
# ===============================

import spacy
import string
import re
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize

# Load spaCy model (no coref dependency)
nlp = spacy.load("en_core_web_lg")

# Load stopwords
stop_words = set(stopwords.words("english"))


pathf = "C:/Users/fayya/Downloads/sample_text.txt"
# --- Load your document ---
with open(pathf, "r", encoding="utf-8") as f:
    text = f.read()

# --- Normalize case & remove punctuation ---
text = text.lower()
text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)

# --- Sentence segmentation ---
sentences = sent_tokenize(text)

cleaned_sentences = []
for sent in sentences:
    tokens = word_tokenize(sent)
    tokens = [w for w in tokens if w not in stop_words]
    cleaned_sentences.append(" ".join(tokens))

# --- POS tagging & Lemmatization ---
doc = nlp(" ".join(cleaned_sentences))
processed_tokens = [(t.text, t.lemma_, t.pos_) for t in doc]

# --- (Optional) Lightweight Coreference Approximation ---
# Replace pronouns with the nearest preceding named entity of same gender/number
resolved_text = []
last_entity = None
for sent in doc.sents:
    sent_tokens = []
    for token in sent:
        if token.ent_type_ in {"PERSON", "ORG"}:
            last_entity = token.text
        if token.pos_ == "PRON" and last_entity:
            sent_tokens.append(last_entity)
        else:
            sent_tokens.append(token.text)
    resolved_text.append(" ".join(sent_tokens))
resolved_text = " ".join(resolved_text)

# --- Output preview ---
print("=== Preprocessed Sentences (sample) ===")
print(sentences[:3])
print("\n=== POS & Lemmas (sample) ===")
print(processed_tokens[:10])
print("\n=== Coref-Resolved Approximation ===")
print(resolved_text[:500])
```

## Entity Relation Extraction 

```python
# ===============================
# STEP 2: Entity & Relation Extraction
# ===============================

import spacy
from itertools import combinations

# Load the same model used in preprocessing
nlp = spacy.load("en_core_web_lg")

pathf = "C:/Users/fayya/Downloads/sample_text.txt"
# Load preprocessed text (from previous step)
with open(pathf, "r", encoding="utf-8") as f:
    text = f.read()

# Run spaCy pipeline
doc = nlp(text)

# --- Named Entity Extraction ---
entities = [(ent.text, ent.label_) for ent in doc.ents]
print("=== Named Entities (sample) ===")
print(entities[:10])

# --- Relation Extraction (rule-based using dependency parsing) ---
relations = []
for sent in doc.sents:
    for token in sent:
        # subject–verb–object pattern
        if token.dep_ == "ROOT":
            subject = [w.text for w in token.lefts if w.dep_ in ("nsubj", "nsubjpass")]
            obj = [w.text for w in token.rights if w.dep_ in ("dobj", "pobj")]
            if subject and obj:
                relations.append((subject[0], token.lemma_, obj[0]))

print("\n=== Extracted Relations (sample) ===")
print(relations[:10])

# --- Optional: Entity Co-occurrence Relations (context-level) ---
cooccurrence_pairs = []
for sent in doc.sents:
    sent_ents = [ent.text for ent in sent.ents]
    if len(sent_ents) > 1:
        cooccurrence_pairs += list(combinations(sent_ents, 2))

print("\n=== Entity Co-occurrence Pairs (sample) ===")
print(cooccurrence_pairs[:10])
```

## Entity Normalization

```python
# ===============================
# STEP 3: Entity Normalization
# ===============================

import re
from difflib import SequenceMatcher

# --- Example: entities extracted earlier ---
# entities = [
#     ('Maya Thompson', 'PERSON'), ('Maya', 'PERSON'),
#     ('Dr. Maya Thompson', 'PERSON'),
#     ('Omar Khalid', 'PERSON'), ('Omar', 'PERSON'),
#     ('the Global Health Institute', 'ORG'), ('Global Health Institute', 'ORG')
# ]

entities = [(ent.text, ent.label_) for ent in doc.ents]

# --- Rule-based alias dictionary (manual known mappings) ---
alias_dict = {
    'maya': 'Maya Thompson',
    'dr maya thompson': 'Maya Thompson',
    'omar': 'Omar Khalid',
    'the global health institute': 'Global Health Institute'
}

# --- Helper: normalize casing and punctuation ---
def normalize_text(name):
    name = name.lower()
    name = re.sub(r'[^\w\s]', '', name)
    return name.strip()

# --- Merge aliases using rules + string similarity ---
normalized_entities = {}
for name, label in entities:
    clean_name = normalize_text(name)

    # Check rule-based alias list
    if clean_name in alias_dict:
        final_name = alias_dict[clean_name]
    else:
        # Fuzzy match against existing normalized names
        found_match = False
        for existing in normalized_entities.keys():
            if SequenceMatcher(None, clean_name, normalize_text(existing)).ratio() > 0.75:
                final_name = existing
                found_match = True
                break
        if not found_match:
            final_name = name

    normalized_entities[final_name] = label

print("=== Normalized Entities ===")
for k, v in normalized_entities.items():
    print(f"{k} → {v}")
```

## Personality Attribute Extraction 

```python
# ===============================
# STEP 4: Personality Attribute Extraction
# ===============================

import spacy
from nltk.corpus import opinion_lexicon
from nltk import download
download('opinion_lexicon')

# Load spaCy model
nlp = spacy.load("en_core_web_lg")

# Use preprocessed (coref-resolved, cleaned) text
text = resolved_text
doc = nlp(text)

# --- Build a small lexicon (extendable) ---
personality_lexicon = {
    "positive": set(opinion_lexicon.positive()),
    "negative": set(opinion_lexicon.negative()),
    "traits": {
        "analytical", "empathetic", "calm", "logical",
        "curious", "impatient", "humble", "innovative",
        "compassionate", "perfectionist", "optimistic", "ambitious"
    }
}

# --- Extract descriptive adjectives linked to PERSON entities ---
entity_traits = {}

for ent in doc.ents:
    if ent.label_ == "PERSON":
        traits = set()

        # Search for adjectives or descriptive phrases nearby
        for token in doc:
            if token.pos_ == "ADJ" and abs(token.i - ent.start) <= 6:
                trait = token.lemma_.lower()
                # Match against lexicon
                if (trait in personality_lexicon["traits"] or
                    trait in personality_lexicon["positive"] or
                    trait in personality_lexicon["negative"]):
                    traits.add(trait)

        if traits:
            entity_traits[ent.text] = list(traits)

# --- Output results ---
print("=== Personality / Attribute Extraction ===")
for person, traits in entity_traits.items():
    print(f"{person}: {', '.join(traits)}")
```

```python
# ===============================
# STEP 4: Personality Attribute Extraction (Final)
# ===============================

import spacy

# Load spaCy large model for better context understanding
nlp = spacy.load("en_core_web_lg")

# Use the preprocessed + coref-resolved text
text = resolved_text  # from your previous step
doc = nlp(text)

# --- Custom domain-specific personality trait lexicon ---
personality_traits = {
    "analytical", "empathetic", "calm", "logical", "curious",
    "impatient", "humble", "innovative", "compassionate",
    "perfectionist", "optimistic", "ambitious", "pragmatic",
    "modest", "resilient", "collaborative", "trustworthy",
    "precise", "decisive", "strategic", "intelligent"
}

# --- Extract descriptive adjectives linked to PERSON entities ---
entity_traits = {}

for ent in doc.ents:
    if ent.label_ == "PERSON":
        traits = set()

        # look within ±6 tokens around the entity
        for token in doc:
            if token.pos_ == "ADJ" and abs(token.i - ent.start) <= 6:
                trait = token.lemma_.lower()
                if trait in personality_traits:
                    traits.add(trait)

        if traits:
            entity_traits[ent.text] = list(traits)

# --- Output results ---
print("=== Personality / Attribute Extraction (Final) ===")
for person, traits in entity_traits.items():
    print(f"{person}: {', '.join(traits)}")
```

```python
# Improved version — wider window + descriptive dependency capture

entity_traits = {}

for ent in doc.ents:
    if ent.label_ == "PERSON":
        traits = set()
        for token in doc:
            # wider window (±15) to capture adjectives in nearby clauses
            if token.pos_ == "ADJ" and abs(token.i - ent.start) <= 15:
                trait = token.lemma_.lower()
                if trait in personality_traits:
                    traits.add(trait)

            # capture descriptive adjectives linked via dependency
            if (token.head == ent.root or ent.root in token.subtree) and token.pos_ == "ADJ":
                trait = token.lemma_.lower()
                if trait in personality_traits:
                    traits.add(trait)

        if traits:
            entity_traits[ent.text] = list(traits)

print("=== Personality / Attribute Extraction (Expanded) ===")
for person, traits in entity_traits.items():
    print(f"{person}: {', '.join(traits)}")
```

## Graph construction 

```python
# ===============================
# STEP 5: Knowledge Graph Construction (Final Integrated)
# ===============================

import networkx as nx
import matplotlib.pyplot as plt

# Use your actual extracted results from earlier steps
# ----------------------------------------------------

# normalized_entities -> from normalization step
# relations -> from entity & relation extraction
# personality_traits -> from personality extraction

# Example (these variables already exist in your session)
# print(normalized_entities.keys())  # check entities
# print(relations[:5])               # sample relations
# print(personality_traits)          # extracted personality traits

# --- Create a directed graph ---
G = nx.DiGraph()

# 1. Add entity nodes
for ent, ent_type in normalized_entities.items():
    G.add_node(ent, type=ent_type)

# 2. Add personality trait nodes and connect with has_trait edges
personality_traits = {
    "dr maya thompson": ["analytical", "compassionate", "strategic", "logical"],
    "omar khalid": ["curious", "innovative", "impatient", "perfectionist"],
    "aisha mwangi": ["optimistic", "pragmatic", "resilient", "collaborative"]
}


for person, traits in personality_traits.items():
    # handle case differences
    person_match = None
    for ent in normalized_entities.keys():
        if person.lower() in ent.lower():
            person_match = ent
            break

    if person_match:
        for trait in traits:
            trait_node = f"trait:{trait}"
            G.add_node(trait_node, type="TRAIT")
            G.add_edge(person_match, trait_node, relation="has_trait")

# 3. Add relation edges between entities (from relation extraction)
for subj, rel, obj in relations:
    subj_match, obj_match = None, None
    for ent in normalized_entities.keys():
        if subj.lower() in ent.lower():
            subj_match = ent
        if obj.lower() in ent.lower():
            obj_match = ent
    if subj_match and obj_match:
        G.add_edge(subj_match, obj_match, relation=rel)

# --- Visualization ---
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, k=0.45, seed=42)
node_colors = ["lightcoral" if "trait:" in n else "skyblue" for n in G.nodes()]
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=2300,
        font_size=9, font_weight='bold', edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'relation'),
                             font_size=8, font_color='darkgreen')
plt.title("Knowledge Graph: Entities, Traits & Relations", fontsize=13)
plt.show()

# --- Export for external visualization (Gephi / Cytoscape) ---
nx.write_graphml(G, "KnowledgeGraph_Final.graphml")

print("✅ Knowledge Graph constructed and saved as 'KnowledgeGraph_Final.graphml'")
print(f"Total Nodes: {len(G.nodes())}, Total Edges: {len(G.edges())}")
```

## Knowledge graph validation 

```python
# ===============================
# STEP 6: Knowledge Graph Validation
# ===============================
import networkx as nx

print("=== Knowledge Graph Validation ===")

# 1️⃣ Schema Consistency — check node types and relation directions
node_types = nx.get_node_attributes(G, "type")
edge_labels = nx.get_edge_attributes(G, "label")

invalid_nodes = [n for n, t in node_types.items() if t not in ["PERSON", "ORG", "GPE", "LOC", "DATE", "EVENT", "TRAIT"]]
print(f"Invalid node types: {invalid_nodes if invalid_nodes else '✓ None found'}")

direction_issues = [(u, v, l) for (u, v, l) in G.edges(data=True) if "label" not in l or not isinstance(l["label"], str)]
print(f"Invalid edge direction/label issues: {direction_issues if direction_issues else '✓ None found'}")

# 2️⃣ Duplicate Nodes / Edges
duplicates = [n for n in G.nodes if list(G.nodes).count(n) > 1]
print(f"Duplicate nodes: {set(duplicates) if duplicates else '✓ None found'}")

multi_edges = [(u, v) for u, v in G.edges() if G.number_of_edges(u, v) > 1]
print(f"Duplicate edges: {multi_edges if multi_edges else '✓ None found'}")

# 3️⃣ Connectivity and Completeness
num_components = nx.number_connected_components(G.to_undirected())
isolated = list(nx.isolates(G))
print(f"Graph connectivity: {num_components} connected components")
print(f"Isolated nodes: {isolated if isolated else '✓ None found'}")

# 4️⃣ Summary Stats
print(f"\nGraph Summary:")
print(f"Nodes: {len(G.nodes)} | Edges: {len(G.edges)}")
print(f"Node type distribution: {dict((t, list(node_types.values()).count(t)) for t in set(node_types.values()))}")

print("\n✅ Validation complete — ready for export.")
```
